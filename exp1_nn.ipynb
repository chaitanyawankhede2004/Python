```python
import numpy as np
import matplotlib.pyplot as plt

# Generate input values
x = np.linspace(-10, 10, 1000)

# Activation functions
def identity(x):
    return x

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def bipolar_sigmoid(x):
    return (2 / (1 + np.exp(-x))) - 1

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def softmax(x):
    exp_x = np.exp(x - np.max(x))
    return exp_x / np.sum(exp_x)

# Create 2x3 subplot layout
fig, axs = plt.subplots(2, 3, figsize=(15, 8))

# Identity
axs[0, 0].plot(x, identity(x), color='blue')
axs[0, 0].set_title("Identity")
axs[0, 0].grid(True)

# Sigmoid
axs[0, 1].plot(x, sigmoid(x), color='green')
axs[0, 1].set_title("Sigmoid")
axs[0, 1].grid(True)

# Bipolar Sigmoid
axs[0, 2].plot(x, bipolar_sigmoid(x), color='red')
axs[0, 2].set_title("Bipolar Sigmoid")
axs[0, 2].grid(True)

# Tanh
axs[1, 0].plot(x, tanh(x), color='purple')
axs[1, 0].set_title("Tanh")
axs[1, 0].grid(True)

# ReLU
axs[1, 1].plot(x, relu(x), color='orange')
axs[1, 1].set_title("ReLU")
axs[1, 1].grid(True)

# Leaky ReLU
axs[1, 2].plot(x, leaky_relu(x), color='brown')
axs[1, 2].set_title("Leaky ReLU")
axs[1, 2].grid(True)

# Common labels
for ax in axs.flat:
    ax.set_xlabel("Input")
    ax.set_ylabel("Output")

plt.tight_layout()
plt.show()

# Softmax in a separate figure
plt.figure(figsize=(6, 4))
plt.plot(x, softmax(x), color='black')
plt.title("Softmax")
plt.xlabel("Input")
plt.ylabel("Probability")
plt.grid(True)
plt.show()
```
